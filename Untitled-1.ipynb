```python
data = {
    "pickup_point": ["Airport", "City"],
    "request_timestamp": [3238, 3507],
    "drop_timestamp": [1327, 1504],
}
df = pd.DataFrame(data)


bar_width = 0.35
x = np.arange(len(df["pickup_point"]))  


plt.bar(x - bar_width / 2, df["request_timestamp"], bar_width, label="Request Timestamp", color="skyblue")
plt.bar(x + bar_width / 2, df["drop_timestamp"], bar_width, label="Drop Timestamp", color="orange")


plt.xlabel("Pickup Point")
plt.ylabel("Timestamp Value")
plt.title("Comparison of Request and Drop Timestamps")
plt.xticks(x, df["pickup_point"])  
plt.legend()


plt.tight_layout()
plt.show()
```

```python




uber['hour'] = uber['request_timestamp'].dt.hour


hourly_bookings = uber.groupby('hour').size().sort_values(ascending=False)

most_bookings_hour = hourly_bookings.idxmax()
least_bookings_hour = hourly_bookings.idxmin()

print(f"Hour with most bookings: {most_bookings_hour} ({hourly_bookings[most_bookings_hour]} bookings)")
print(f"Hour with least bookings: {least_bookings_hour} ({hourly_bookings[least_bookings_hour]} bookings)")


plt.figure(figsize=(12, 6))
hourly_bookings.sort_index().plot(kind='bar', color='skyblue', edgecolor='black')


plt.xlabel("Hour of the Day")
plt.ylabel("Number of Bookings")
plt.title("Hourly Distribution of Uber Bookings")
plt.xticks(rotation=0)

plt.text(most_bookings_hour, hourly_bookings[most_bookings_hour] + 1,
         f"Most: {hourly_bookings[most_bookings_hour]}", color='green', ha='center')
plt.text(least_bookings_hour, hourly_bookings[least_bookings_hour] + 1,
         f"Least: {hourly_bookings[least_bookings_hour]}", color='red', ha='center')

plt.tight_layout()
plt.show()
```

```python
uber['hour'] = uber['request_timestamp'].dt.hour


cancelled_bookings = uber[uber['status'] == 'Cancelled']


hourly_cancellations = cancelled_bookings.groupby('hour').size().sort_values(ascending=False)


most_cancellations_hour = hourly_cancellations.idxmax()
least_cancellations_hour = hourly_cancellations.idxmin()

print(f"Hour with most cancellations: {most_cancellations_hour} ({hourly_cancellations[most_cancellations_hour]} cancellations)")
print(f"Hour with least cancellations: {least_cancellations_hour} ({hourly_cancellations[least_cancellations_hour]} cancellations)")


plt.figure(figsize=(12, 6))
hourly_cancellations.sort_index().plot(kind='bar', color='salmon', edgecolor='black')


plt.xlabel("Hour of the Day")
plt.ylabel("Number of Cancellations")
plt.title("Hourly Distribution of Uber Cancellations")
plt.xticks(rotation=0)


plt.text(most_cancellations_hour, hourly_cancellations[most_cancellations_hour] + 1,
         f"Most: {hourly_cancellations[most_cancellations_hour]}", color='green', ha='center')
plt.text(least_cancellations_hour, hourly_cancellations[least_cancellations_hour] + 1,
         f"Least: {hourly_cancellations[least_cancellations_hour]}", color='red', ha='center')

plt.tight_layout()
plt.show()
```

```python
no_cars_available = uber[uber['status'] == 'No Cars Available']


hourly_no_cars = no_cars_available.groupby('hour').size().sort_values(ascending=False)


most_no_cars_hour = hourly_no_cars.idxmax()
least_no_cars_hour = hourly_no_cars.idxmin()

print(f"Hour with most 'No Cars Available': {most_no_cars_hour} ({hourly_no_cars[most_no_cars_hour]} occurrences)")
print(f"Hour with least 'No Cars Available': {least_no_cars_hour} ({hourly_no_cars[least_no_cars_hour]} occurrences)")

plt.figure(figsize=(12, 6))
hourly_no_cars.sort_index().plot(kind='bar', color='orange', edgecolor='black')


plt.xlabel("Hour of the Day")
plt.ylabel("Number of 'No Cars Available' Occurrences")
plt.title("Hourly Distribution of 'No Cars Available'")
plt.xticks(rotation=0)


plt.text(most_no_cars_hour, hourly_no_cars[most_no_cars_hour] + 1,
         f"Most: {hourly_no_cars[most_no_cars_hour]}", color='green', ha='center')
plt.text(least_no_cars_hour, hourly_no_cars[least_no_cars_hour] + 1,
         f"Least: {hourly_no_cars[least_no_cars_hour]}", color='red', ha='center')

plt.tight_layout()
plt.show()
```

```python



no_cars_available = uber[uber['status'] == 'No Cars Available']


grouped_no_cars = no_cars_available.groupby(['pickup_point', 'hour']).size().sort_values(ascending=False)

print("Top 10 Pickup Point and Hour combinations with 'No Cars Available':")
print(grouped_no_cars.head(10))

grouped_no_cars = grouped_no_cars.reset_index(name='count')


pivot = grouped_no_cars.pivot(index='hour', columns='pickup_point', values='count').fillna(0)


plt.figure(figsize=(12, 8))
pivot.plot(kind='bar', stacked=True, figsize=(12, 6), color=['skyblue', 'salmon'], edgecolor='black')


plt.xlabel("Hour of the Day")
plt.ylabel("Number of 'No Cars Available'")
plt.title("'No Cars Available' by Pickup Point and Hour")
plt.xticks(rotation=0)
plt.legend(title="Pickup Point", loc='upper right')

plt.tight_layout()
plt.show()
```

```python
cancelled_bookings = uber[uber['status'] == 'Cancelled']


grouped_cancellations = cancelled_bookings.groupby(['pickup_point', 'hour']).size().sort_values(ascending=False)

print("Top 10 Pickup Point and Hour combinations with Cancellations:")
print(grouped_cancellations.head(10))


grouped_cancellations = grouped_cancellations.reset_index(name='count')


pivot = grouped_cancellations.pivot(index='hour', columns='pickup_point', values='count').fillna(0)


plt.figure(figsize=(12, 8))
pivot.plot(kind='bar', stacked=True, figsize=(12, 6), color=['skyblue', 'salmon'], edgecolor='black')


plt.xlabel("Hour of the Day")
plt.ylabel("Number of Cancellations")
plt.title("Cancellations by Pickup Point and Hour")
plt.xticks(rotation=0)
plt.legend(title="Pickup Point", loc='upper right')

plt.tight_layout()
plt.show()
```
